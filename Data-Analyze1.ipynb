# Generating values for null

## Model based Imputation for Numerical data

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=8)
numerical_columns = ['Age','Annual Income','Number of Dependents','Health Score','Previous Claims','Vehicle Age','Credit Score','Insurance Duration']
train_data[numerical_columns] = imputer.fit_transform(train_data[numerical_columns])
```

## Label Encoding + KNN Imputation for Categorical data

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = ['Marital Status', 'Occupation']
encoder = LabelEncoder()

for column in categorical_columns:
    train_data[column] = encoder.fit_transform(train_data[column].astype(str))

imputer = KNNImputer(n_neighbors=5)
train_data[categorical_columns] = imputer.fit_transform(train_data[categorical_columns])

# Revert encoding if necessary
for column in categorical_columns:
    train_data[column] = encoder.inverse_transform(train_data[column].astype(int))
```
